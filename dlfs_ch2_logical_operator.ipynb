1), 2), 3)은 단층 퍼셉트론으로 구현

4)는 다층 퍼셉트론을 통해 구현



```
w1*x1 + w2*x2 = y        # if y >= theta : true
                         # if y <  theta : false
```


```
def LOperator(x1, x2):
    x = np.array([x1, x2])    # input
    w = np.array([w1, w2])    # weight : 매개변수
    b = -theta                # bias   : 매개변수
    tmp = np.sum(w*x) + b     # output
    if tmp <= 0:
        return 0
    else:
        return 1
```





### ex) AND GATE

````python
"""
example)
w1 = 0.5
w2 = 0.5
theta = 0.7 일 때

0.5 * x1 + 0.5 * x2 = y (if y >= 0.7 : true
                         if y <  0.7 : false )
=> sum([x1, x2]*[0.5, 0.5]) - 0.7 = y (if y >  0 : true
                                       if y <= 0 : false )
=> sum(x * y) + b = tmp
"""
# coding: utf-8
import numpy as np


def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7 # bias = - theta
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = AND(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
````

### ex) NAND GATE

````python
"""
example)
w1 = -0.5
w2 = -0.5
theta = -0.7 일 때

-0.5 * x1 - 0.5 * x2 = y (if y >= -0.7 : true
                          if y <  -0.7 : false )
=> sum([x1, x2]*[-0.5, -0.5]) + 0.7 = y (if y >  0 : true
                                         if y <= 0 : false )
=> sum(x * y) + b = tmp
"""

def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = NAND(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
````

### ex) OR GATE

````python
"""
example)
w1 = 0.5
w2 = 0.5
theta = 0.2 일 때

0.5 * x1 + 0.5 * x2 = y  (if y >= 0.2 : true
                          if y <  0.2 : false )
=> sum([x1, x2]*[0.5, 0.5]) - 0.2 = y   (if y >  0 : true
                                         if y <= 0 : false )
=> sum(x * y) + b = tmp
"""

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = OR(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
````

### ex) NOR GATE

````python
# coding: utf-8

""" 다층 퍼셉트론을 통해 구현 """
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

if __name__ == '__main__':
    for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
        y = XOR(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
````
