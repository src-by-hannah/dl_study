```python
# 출력을 위한 import
import numpy as np
import matplotlib.pylab as plt
```

```python
# 계단 함수
def step_function(x) :
    y = x > 0 # x : numpy array
    return y.astype(np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)

print(y)

plt.style.use(['dark_background'])
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 시그모이드 함수
def sigmoid(x) :
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

print(y)

plt.style.use(['dark_background'])
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# ReLU 함수
def relu(x) :
    return np.maximum(0, x)
    
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

print(y)

plt.style.use(['dark_background'])
plt.plot(x, y)
plt.show()
```

```python
# 소프트맥스 함수
def softmax(a) : # a = w1*x1 + ... + wn*xn + b
    c = np.max(a) # overflow 방지를 위함
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y

a = np.array([1010, 1000, 900])
y = softmax(a)

print(y)

print(np.sum(y)) # softmax 함수 결과값의 합은 1임
```
